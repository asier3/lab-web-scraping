````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
response = requests.get(url=url)
response.status_code
````

````python
soup = BeautifulSoup(response.content)
movies = soup.find_all('td', attrs={'class':'titleColumn'})
len(movies)
````

````python
# We can get text this way
# Somehow i cannot change the language to english and the movie title only appears in portuguese lol...
import re

title_date = movies[0].text.strip().replace('\n', ',')
title_date = re.sub('[\d]+\.\,','', title_date).strip().split(',')
title_date
````

````python
# We can find the director and stars in the following part of HTML, and append it to our list above

dir_stars = movies[0].a.get('title')
dir_stars
````

````python
def strip_titles(element):
    element = element.text.strip().replace('\n', ',')
    element = re.sub('[\d]+\.\,','', element).strip().split(',')
    return element

titles_list = list(map(strip_titles,movies))

#titles_df = pd.DataFrame(titles_list, columns={'release','title','nones'})   ---> (why the columns are assorted in a random order?)
titles_df = pd.DataFrame(titles_list, columns=('title','release','nones'))
titles_df.head(4)
````

````python
# Combining both dataframes we will get the answer we are looking for

titles_df['director_and_stars'] = stars_df['director_and_stars']
titles_df. drop('nones', axis=1, inplace=True)
titles_df

# I am pretty sure there's a faster way of extracting all the info at once from the HTML but couldn't really come up with it...
# Also, as I said, I cannot change the language in the website and as a result I get the titles in portuguese...
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.
